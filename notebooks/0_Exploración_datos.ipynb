# Exploración de datasets

## 0. Setup del proyecto

```python
import pandas as pd

ruta_inegi = "../data/raw/censo_inegi_2020.csv"
```

## 1. Dataset INEGI de Censo poblacional 2020

```python
#Cargamos el dataframe
df_inegi = pd.read_csv(ruta_inegi, dtype=str)

#Imprimimos tamaño para ver tamaño original
print(f"Tamaño original: {df_inegi.shape[0]} filas y {df_inegi.shape[1]} columnas.")

# Vemos las primeras 5 filas para entender la estructura
display(df_inegi.head())
```

### 1.1. Filtramos por la capital

```python
# Filtramos usando la columna 'MUN' (Municipio)
# Usamos zfill(3) por si Pandas leyó un '30' en lugar de '030'
df_hmo = df_inegi[df_inegi['MUN'].str.zfill(3) == '030'].copy()

#Imprimimos tamaño para ver tamaño modificado
print(f"Tamaño Hermosillo: {df_hmo.shape[0]} filas y {df_hmo.shape[1]} columnas.")
df_hmo.head()
```

```python
# Buscar columnas que contengan cierta palabra clave (ej. 'VIV' para viviendas)
palabra_clave = 'VIV'
columnas_encontradas = [col for col in df_hmo.columns if palabra_clave in col]

print(f"Columnas que contienen '{palabra_clave}':")
print(columnas_encontradas)

# Prueba cambiando la palabra clave a 'DREN' o 'INTER'
```
